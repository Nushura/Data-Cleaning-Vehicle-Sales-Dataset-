# Task 1: Data Preparation


```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame

# these are imported for visualization with graph
import matplotlib.pyplot as plt
from matplotlib import rcParams

import seaborn as sns
# %matplotlib inline
```

```python
# Retrieving data files from computer files
manufacturer_data = 'C:/Users/Nushura/Documents/S4/Practical Data Science with Python/Assignments/A1/Manufacturer.csv'
manufacturer = pd.read_csv(manufacturer_data)

model_data = 'C:/Users/Nushura/Documents/S4/Practical Data Science with Python/Assignments/A1/Model.csv'
model = pd.read_csv(model_data)

price_data = 'C:/Users/Nushura/Documents/S4/Practical Data Science with Python/Assignments/A1/Price.csv'
price = pd.read_csv(price_data)

transmission_data = 'C:/Users/Nushura/Documents/S4/Practical Data Science with Python/Assignments/A1/Transmission.csv'
transmission = pd.read_csv(transmission_data)

power_data = 'C:/Users/Nushura/Documents/S4/Practical Data Science with Python/Assignments/A1/Power.csv'
power = pd.read_csv(power_data)

engine_data = 'C:/Users/Nushura/Documents/S4/Practical Data Science with Python/Assignments/A1/Engine CC.csv'
engine = pd.read_csv(engine_data)

fuel_data = 'C:/Users/Nushura/Documents/S4/Practical Data Science with Python/Assignments/A1/Fuel.csv'
fuel = pd.read_csv(fuel_data)

male_data = 'C:/Users/Nushura/Documents/S4/Practical Data Science with Python/Assignments/A1/Male.csv'
male = pd.read_csv(male_data)

female_data = 'C:/Users/Nushura/Documents/S4/Practical Data Science with Python/Assignments/A1/Female.csv'
female = pd.read_csv(female_data)

unknown_data = 'C:/Users/Nushura/Documents/S4/Practical Data Science with Python/Assignments/A1/Unknown.csv'
unknown = pd.read_csv(unknown_data)

total_data = 'C:/Users/Nushura/Documents/S4/Practical Data Science with Python/Assignments/A1/Total.csv'
total = pd.read_csv(total_data)
```

```python
# Concatenating data from the retirieved files into one single dataframe
# Concatination is being based on row index values

x = pd.concat([manufacturer, model, price, transmission, power, engine, fuel, male, female, unknown, total], axis=1)
x.drop("Unnamed: 0", axis=1)
```

```python
# Filtering out missing values

# this will drop all rows from the table that has any missing values
# by using dropna() method

new_dataset_without_missing_values = x.dropna()
new_dataset_without_missing_values

# After doing this it can be seen that the number of rows decreased from 6102 to 6097
# As it is a alarge dataset and after dropping the missing values rows the difference was not too much
# Thats why this will not be a problem as there are a lot more data present after removal of missing values rows
```

```python
# Checking for duplicated rows
new_dataset_without_missing_values.duplicated()
new_dataset_without_missing_values.drop_duplicates()

# After trying to drop the duplicated rows it can be observed that no rows were dropped
# Thus, in can be understood that there were no duplicated dataset rows
```

```python
df = new_dataset_without_missing_values

# Converting all the columns from object type to string type 
# followd by removal of commas and conversion from string type to integer type

df['Total'].astype(str)
df['Total'] = df['Total'].str.replace(',', '').astype(int)

df['Male'].astype(str)
df['Male'] = df['Male'].str.replace(',', '').astype(int)

df['Female'].astype(str)
df['Female'] = df['Female'].str.replace(',', '').astype(int)

df['Unknown'].astype(str)
df['Unknown'] = df['Unknown'].str.replace(',', '').astype(int)

df.nlargest(6096, 'Total', keep='first')

# Removing the Unnamed column which acted like an index for each table
y = df.drop("Unnamed: 0", axis=1)
# Creating the final cleaned_car_buyers csv file
y.to_csv('cleaned_car_buyers.csv', index=False)

```

## Price Boxplot

```python
# Price
# Initial pre-cleaned box plot of price
# using the seaborn library
sns.boxplot(new_df['Price'])
```

```python
# Styling
sns.set(style='white')

# Setting figure size
fig, ax = plt.subplots(figsize=(10, 2))

# Boxplot of Price column
# Setting the color to be plum and width 
sns.boxplot(new_df['Price'], color="plum", width=.6)

# Labeling of title and x-axis
plt.title('Price Boxplot', fontsize=14)
plt.xlabel('Price')

# This is to remove the borders around the boxplot
sns.despine()

# Setting x limit
plt.xlim([0, 700])

plt.show()
```

```python
# As the initial box plot was difficult to analyse 
# this zoomed in box plot is created for price 
# and it can be observed that any value above 80 is an outlier
sns.set(style='white')

fig, ax = plt.subplots(figsize=(10, 2))

sns.boxplot(new_df['Price'], color="plum", width=.6)

plt.title('Price Boxplot', fontsize=13)
plt.xlabel('Price')

# to remove the border around the box plot graph
sns.despine()

# Changed limit of x-axis to zoom into the boxplot 
# to observe the point from after which the outliers started
plt.xlim([0, 100])

plt.show()
```

```python
# This code is enabling all the values before 1.6 and after 82 to be removed
# As those values are acting as outliers
# Thus, after removing those values, the new dataset is being stored in 'k'
k = new_df[new_df['Price'].between(1.6,82)]
```

```python
sns.set(style='white')

fig, ax = plt.subplots(figsize=(10, 2))

sns.boxplot(k['Price'], color="plum", width=.6)

plt.title('Price Boxplot', fontsize=13)
plt.xlabel('Price')

# to remove the border around the box plot graph
sns.despine()

plt.xlim([0, 100])

plt.show()
```

```python
# This code is enabling all the values before 1.6 and after 66.6 to be removed from price column
# As those values are acting as outliers
# Thus, after removing those values, the new dataset is being stored in 'new_df_no_Price_Outliers'
new_df_no_Price_Outliers = new_df[new_df['Price'].between(1.6,66.6)]
```

```python
sns.set(style='white')

fig, ax = plt.subplots(figsize=(10, 2))

sns.boxplot(new_df_no_Price_Outliers['Price'], color="plum", width=.6)

plt.title('Price Boxplot', fontsize=13)
plt.xlabel('Price')

# to remove the border around the box plot graph
sns.despine()

plt.xlim([0, 100])

plt.show()
```
